### Exercici 3
__Crea una matriu bidimensional 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.__

```python
# Matriu 5x5
matriu_2 = np.array([[3,5,1,7,8],[5,1,7,5,9],[3,6,1,7,8],[1,8,9,2,3],[5,4,6,7,2]])
print(matriu_2)
print("La dimensió de la matriu és: " + str(matriu_2.ndim))
print("La forma de la matriu és: " + str(matriu_2.shape))
```

```python
# Màxim de la matriu
matriu_max = matriu_2.max()
print("El valor màxim de la matriu és: " + str(matriu_max))
```

```python
# Valor màxim per files i columnes
def maxim_eix(matriu):
    for i in range(matriu.shape[0]):
        print("El valor màxim de la fila " + str(i) + " és: " + str(matriu[i,:].max()))
    for j in range(matriu.shape[1]):
        print("El valor màxim de la columna " + str(j) + " és: " + str(matriu[:,j].max()))

maxim_eix(matriu_2)
```

### Exercici 4
__Mostra amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu: "les matrius es poden transmetre (broadcast) entre sí si les seves dimensions coincideixen o si una de les matrius té una mida d'1".__

```python
# Matriu i escalar
cons = np.array(3) # Un escalar
matriu_a = np.array([1,2]) # 1 dimensió
matriu_cons = np.array([3,3])
print(matriu_a+matriu_cons)
print(matriu_a+cons)
```

<span style='color:blue'> Operar dues matrius de la mateixa dimensió (una d'elles amb el mateix valor repetit) és el mateix que operar una matriu amb un escalar. S'ha produït un broadcasting per a l'escalar. </span>

```python
# Matriu d'1 dimensió i matriu de dos dimensions
matriu_a = np.array([1,2]) # 1 dimensió
matriu_b = np.array([[5,6,7],[5,2,4]]) # 2x3
print(matriu_a)
print('+')
print(matriu_b)
print(matriu_a+matriu_b)
```

<span style='color:blue'> No es pot realitzar l'operació, ja que les dimensions no coincideixen (la matriu a té 1 dimensió y la matriu b té 2). </span>

```python
# Es redimensiona la matriu a amb dues files i una columna
matriu_a_2 = np.array([[1,1,1],[2,2,2]])
print(matriu_a_2)
print('+')
print(matriu_b)
print('---------')
print(matriu_a_2+matriu_b)
```

```python
# Broadcast de la matriu a
matriu_a_3 = matriu_a.reshape(2,1)
print(matriu_a_3)
print('+')
print(matriu_b)
print('---------')
print(matriu_a_3+matriu_b)
```

<span style='color:blue'> Com que ara les dues matrius tenen la mateixa dimensió, i el nombre de files coincideix, llavors es produeix un broadcasting i es pot realitzar l'operació: a la fila 1 de la matriu b es suma 1, i a la fila 2 de la matriu b es suma 2.</span>

```python
# Matriu d'1 dimensió i matriu de tres dimensions
matriu_a = np.array([1,2]) # 1 dimensió
matriu_c = np.array([[[5,6,1],[5,2,5]],[[4,1,2],[5,3,1]]]) # 2x2x2
print(matriu_a)
print('+')
print(matriu_c)
print(matriu_a+matriu_c)
```

<span style='color:blue'> No es pot realitzar l'operació, ja que les dimensions no coincideixen (la matriu a té 1 dimensió y la matriu c té 3). </span>

```python
# Es redimensiona la matriu a amb la forma 2x1x1
matriu_a_4 = np.array([[[1,1,1],[2,2,2]],[[1,1,1],[2,2,2]]])
print(matriu_a_4)
print('+')
print(matriu_c)
print('---------')
print(matriu_a_4+matriu_c)
```

```python
# Broadcast de la matriu a
matriu_a_5 = matriu_a.reshape(2,1,1)
print(matriu_a_5)
print('+')
print(matriu_c)
print('---------')
print(matriu_a_3+matriu_c)
```

<span style='color:blue'> Com que ara les dues matrius tenen la mateixa dimensió, llavors es produeix un broadcasting i es pot realitzar l'operació.</span>

### Exercici 5
__Utilitza la Indexació per extreure els valors d'una columna i una fila de la matriu i suma els seus valors.__

```python
matriu_3 = np.array([[3,4,1,7],[5,1,5,9],[6,1,7,8],[1,8,9,3],[5,6,7,2]])

def suma_fila_col(matriu,fila,columna):
    try:
        f = matriu[fila,:]
        c = matriu[:,columna]
        res_fila = f.sum()
        res_col = c.sum()
        resultat = res_fila + res_col
        print(matriu)
        print('--------')
        print('La suma de la fila '+ str(f) + ' és: ' + str(res_fila))
        print('La suma de la columna '+ str(c) + ' és: ' + str(res_col))
        print('La suma de la fila i la columna és: ' + str(resultat))
    except IndexError:
        print('Error: algun index no existeix en la matriu')
    
suma_fila_col(matriu_3,1,1)
```

### Exercici 6
__Mask la matriu anterior, realitza un càlcul booleà vectoritzat, agafant cada element i comprovant si es divideix uniformement per quatre.
Això retorna una matriu de mask de la mateixa forma amb els resultats elementals del càlcul.__

```python
# S'enmascara la matriu pels valors divisibles per 4
matriu_3 = np.array([[3,4,1,7],[5,8,5,9],[6,1,7,8],[1,8,9,3],[5,6,7,4]])
matriu_mask = ma.masked_where(matriu_3 % 4 == 0, matriu_3)
print(matriu_mask) # Matriu amb els divisibles per 4 enmascarats
print('-------')
print(matriu_mask.mask) # Matriu booleana, on TRUE indica els valors divisibles per 4
```
